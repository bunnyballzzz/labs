```python
import random
import matplotlib.pyplot as plt
import math as math
%matplotlib inline
```

task1

```python
x = list(range(-10,11,1))
print(x)
x = list(map(lambda el: el/30,x))
y = list(map(lambda el: 2 * math.sin(3*el*math.pi),x))
print(x)
plt.plot(x,y)
plt.show()
```
